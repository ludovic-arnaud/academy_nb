# Analysez des données de systèmes éducatifs

Academy, une start-up de la EdTech, propose des contenus de formation en ligne pour un public de niveau lycée et université.
Elle cherche à étendre son activité à l’international.
Pour se faire, on nous demande d’étudier les données sur l’éducation de la World Bank EdStats All Indicator Query. 
L’exploitation de ces données va permettre de réaliser un top des pays et des régions où Academy va pouvoir se concentrer pour se développer.

La World Bank EdStats All Indicator Query répertorie plus de 4000 indicateurs internationaux décrivant les indicateurs de l’éducation qui sont comparables à l’échelle internationale et qui décrivent l’accès à l’éducation, la progression, l’achèvement, l’alphabétisation, les enseignants, la population et les dépenses.
Les indicateurs couvrent le cycle de l’éducation de la prématernelle à l’enseignement professionnel et supérieur. 
La requête contient également des données sur les résultats d’apprentissage provenant d’évaluations internationales et régionales, des données sur l’équité provenant d’enquêtes auprès des ménages et des données de projection/atteinte jusqu’en 2050.

```python
def get_null_heatmap(df):
    """ Get a visualization of null values in a dataframe"""
    pyplot.figure(figsize=(20, 10))
    sns.heatmap(df.isna(), cbar=False)
    pyplot.title("Valeurs nulles parc colonne")
    pyplot.ylabel("Nombre de valeurs nulles")
    pyplot.show()


get_null_heatmap(country)
```

```python
def get_duplicates(df):
    """Identify the columns with duplicates and show the number of duplicates for each of them for a dataframe"""
    columns = df.columns.tolist()
    for column in columns:
        duplicates = df[df.duplicated([column], keep=False)]
        if duplicates.shape[0] != 0:
            print("Nombre de doublons sur l'indicateur {} : {}".format(
                column, duplicates.shape[0]))


get_duplicates(country)
```

```python
# non null row counts per column for years between 1970 and 2100
counts = first_selection_copy.loc[:, '1970':'2100'].notnull().sum()
counts.head()
```

```python
def create_plot(df, color, title, ylabel):
    # create matplotlib figure with 20" width and 10" height
    pyplot.figure(figsize=(20, 10))
    # create grid 1 column, 1 row
    grid = gridspec.GridSpec(1, 1)
    # bar style counts
    graph = df.plot(kind='bar', color=color)
    graph.set_title(title)
    graph.set_ylabel(ylabel)
    graph.grid(False)


create_plot(counts, 'SteelBlue',
            'Nombre de données non nulles par année', 'Nombre de données')
```

## 3. Analyse exploratoire des données

### 3.1. Internet users IT.NET.USER.P2

Moyenne, écart type et médiane de 2005 à 2016 par pays

```python
internet_data = final_selection.loc[final_selection['Indicator Code']
                                    == 'IT.NET.USER.P2']
```

```python
def get_mean_std_median(df):
    """Calculate mean, std and median and add it to dataframe"""
    df_copy = df.copy()
    df_copy['Mean'] = df_copy.loc[:, '2005':'2016'].mean(
        axis=1, numeric_only=True)
    df_copy['STD'] = df_copy.loc[:, '2005':'2016'].std(
        axis=1, numeric_only=True)
    df_copy['Median'] = df_copy.loc[:, '2005':'2016'].median(
        axis=1, numeric_only=True)
    return df_copy


internet_data_copy = get_mean_std_median(internet_data)
internet_data_copy.head()
```

L'indicateur IT.NET.USER.P2 est une mesure de la proportion d'utilisateurs d'internet par an. Cette mesure augmente d'année en année. Prendre la moyenne de 2005 à 2016 pour la suite de notre étude n'est pas judicieux. On le voit avec les valeurs d'écart-type qui peuvent être élevés pour certains pays.

Voyons un peu la distribution des données en 2016 :

```python
internet_data_copy.boxplot(column="2016", vert=False)
pyplot.title("Distribution des données internet en 2016")
pyplot.xlabel("Proportion")
pyplot.show()
```

Je vais prendre les proportions de 2016 pour la suite de l'étude sur l'indicateur internet par pays.

```python
sorted_internet = internet_data_copy.sort_values(by=['2016'], ascending=False)
sns.set(style="whitegrid")
pyplot.figure(figsize=(15, 100))
pyplot.title('Proportion par Pays en 2016', size=15)
sns.barplot(data=sorted_internet, x="2016", y="Country Name")
pyplot.show()
```

Les 10 pays ayant la plus grande proportion d'utilisateurs d'internet sont :

```python
internet_data_copy.sort_values(by=['2016'], ascending=False).head(10)
```

Pour les régions je vais faire une moyenne des proportions de 2016 pour les pays de chaque région.

```python
internet_by_group = internet_data_copy.groupby(
    'Region')['2016'].mean().reset_index()
internet_by_group.head()
```

Voici une répartition globale par région

```python
pyplot.pie(x=internet_by_group['2016'],
           labels=internet_by_group['Region'], autopct='%.2f%%')
```

```python
sorted_internet_by_group = internet_by_group.sort_values(
    by=['2016'], ascending=False)
sns.set(style="whitegrid")
pyplot.figure(figsize=(15, 10))
pyplot.title('Proportion par Région en 2016', size=15)
sns.barplot(data=sorted_internet_by_group, x="2016", y="Region")
pyplot.show()
```

Les 3 régions ayant la plus grande proportion d'utilisateurs d'internet sont :

```python
internet_by_group.sort_values(by=['2016'], ascending=False).head(3)
```

### 3.4. Croisement des données internet et lycéens/étudiants

Pour déterminer le top 5 des pays où Academy peut développer son activité, je vais croiser les données des utilisateurs internet et du nombre de lycéens/étudiants. Pour cela, je vais appliquer la proportions d'utilisateurs internet sur la population de lycéens/étudiants.

```python
# select Country Name, Region and 2016 from IT.NET.USER.P2
internet_selection = internet_data[["Country Name", "Region", "2016"]]
```

```python
# select Country Name and Mean from UIS.E.3 & SE.TER.ENRL
education_selection = education_data_copy[["Country Name", "Mean"]]
```

```python
internet_education = internet_selection.merge(
    education_selection, left_on='Country Name', right_on='Country Name')
internet_education["Students with Internet"] = (
    internet_education["2016"] * internet_education["Mean"])/100
internet_education.rename(columns={"2016": "Internet", "Mean": "Education"})
```

```python
sorted_internet_education = internet_education.sort_values(
    by=['Students with Internet'], ascending=False).head(10)
sns.set(style="whitegrid")
pyplot.figure(figsize=(15, 10))
pyplot.title(
    "Nombre de lycéens et d'étudiants moyen par Pays entre 2005 et 2016 ayant Internet", size=15)
sns.barplot(data=sorted_internet_education,
            x="Students with Internet", y="Country Name")
pyplot.show()
```

On peut établir le même type de classement par région géographique

```python
internet_education_by_group = internet_education.groupby(
    'Region')['Students with Internet'].sum().reset_index()
```

```python
pyplot.pie(x=internet_education_by_group['Students with Internet'],
           labels=internet_education_by_group['Region'], autopct='%.2f%%')
```

```python
sorted_internet_education_by_group = internet_education_by_group.sort_values(
    by=['Students with Internet'], ascending=False)
sns.set(style="whitegrid")
pyplot.figure(figsize=(15, 10))
pyplot.title(
    "Nombre de lycéens et d'étudiants moyens par Région entre 2005 et 2016 ayant Internet", size=15)
sns.barplot(data=sorted_internet_education_by_group,
            x="Students with Internet", y="Region")
pyplot.show()
```
